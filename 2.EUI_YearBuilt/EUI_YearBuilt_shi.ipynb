```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
berdo = pd.read_csv("berdo2017.csv",encoding='latin-1')
#pa = pd.read_csv("property-assessment-fy2017.csv",encoding='latin-1')
#berdo_pa = pd.read_csv("together.csv",encoding='latin-1')
```

```python
year_eui = berdo.loc[:,["Property Type","Year Built","Site EUI (kBTU/sf)"]]
before = len(year_eui)
year_eui = year_eui[year_eui["Site EUI (kBTU/sf)"]!="Not Available"]
after = len(year_eui)
print("Buildings with unavailable Site EUI have been removed . \nThe total number of building data reduced from {} to {}.".format(before,after))
```

```python
to_drop = [37, 47, 81, 493, 501, 942, 1001, 1056, 1329, 1521, 1658, 1794]
to_drop += [274, 877, 1181, 1631, 1259, 1420]
year_eui = year_eui.drop(to_drop)
year_eui
```

```python
##################https://stackoverflow.com/questions/15891038/change-data-type-of-columns-in-pandas###################
year_eui[['Site EUI (kBTU/sf)']] = year_eui[['Site EUI (kBTU/sf)']].apply(pd.to_numeric)
```

```python
#year_eui_without_outliers.to_csv("year_eui_without_outliers.csv")
```

```python
year_eui.sort_values(by="Year Built",ascending=True)
```

```python
def draw(str="All Types"):
    if str == "All Types":
        type = year_eui
    else:
        type = year_eui[year_eui["Property Type"]==str]
    year = []
    eui = []
    for i in type.iterrows():
        year.append(i[1][1])
        eui.append(i[1][2])
    plt.figure(figsize=(10,5))
    plt.plot(year, eui,"ro",markersize=3,c="g") 
    plt.title(str)
    plt.xlabel('Built Year')
    plt.ylabel('EUI')
    plt.show()
```

```python
draw()
```

```python
# draw main types of building's yearbuilt-EUI plot
###################https://stackoverflow.com/questions/20076195/what-is-the-most-efficient-way-of-counting-occurrences-in-pandas#########################
top_types = list(year_eui['Property Type'].value_counts()[:20].index)
for type in top_types:
    draw(type)
```
