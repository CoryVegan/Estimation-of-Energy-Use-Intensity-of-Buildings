# Average EUI for each property type 2017

2.the integrated form, too.

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
berdo = pd.read_csv("berdo2017.csv",encoding='latin-1')
#pa = pd.read_csv("property-assessment-fy2017.csv",encoding='latin-1')
#berdo_pa = pd.read_csv("together.csv",encoding='latin-1')
```

```python
type_eui = berdo.loc[:,["Property Type","Site EUI (kBTU/sf)"]]
before = len(type_eui)
type_eui = type_eui[type_eui["Site EUI (kBTU/sf)"]!="Not Available"]
after = len(type_eui)
print("Buildings with unavailable Site EUI have been removed . \nThe total number of building data reduced from {} to {}.".format(before,after))
type_eui
```

```python
# change the type of Site EUI to float
for i in range(len(type_eui)):
    type_eui.iloc[i][1] = float(type_eui.iloc[i][1])
```

```python
type_eui_count = type_eui.copy()
type_eui_count["count"] = np.ones(len(type_eui),dtype=int)
```

```python
# The top most property types
count = type_eui_count.groupby("Property Type").sum()
count.sort_values(by="count",ascending=False)
```

```python
type_eui.sort_values(by='Site EUI (kBTU/sf)',ascending=False)
```

```python
# According to the above form,drop those outliers with a EUI greater than 800
to_drop = []
for key,value in type_eui.iterrows():
    if value[1]>800:
        to_drop.append(key)
type_eui = type_eui.drop(to_drop)
try:type_eui = type_eui.drop([1521])
except:pass
type_eui.sort_values(by='Site EUI (kBTU/sf)', ascending=False)
```

```python
eui = type_eui.groupby("Property Type").sum()
eui_count = eui.merge(count,left_index=True,right_index=True)
ave_eui =[]
for key,value in eui_count.iterrows():
    tmp = value[0]/value[1]
    ave_eui.append(tmp)
type_ave_eui = eui_count.copy()
type_ave_eui["ave_eui"] = ave_eui
type_ave_eui = type_ave_eui.loc[:,["ave_eui"]]
type_ave_eui = type_ave_eui.sort_values(by='ave_eui', ascending=False)
type_ave_eui
```

```python
type_ave_eui.plot.bar(figsize=(18,5),title="Average EUI for each property type")
```

# Average EUI for Main Types

```python
threshold = 5
# determine the drop list
t = count[count["count"]<threshold]
drop_list = list(t.index)
count_dropped = count.drop(drop_list)
print("length of drop list is ",len(drop_list))

# drop those types with less than x buildings in total
type_eui_dropped = type_eui.copy()
for drop in drop_list:
    tmp1 = list(type_eui_count[type_eui_count["Property Type"] == drop].index)
    try:
        type_eui_dropped = type_eui.drop(tmp1)
    except:
        print("error with dropping ",drop)
```

```python
eui = type_eui_dropped.groupby("Property Type").sum()
eui_count = eui.merge(count_dropped,left_index=True,right_index=True)
ave_eui =[]
for key,value in eui_count.iterrows():
    tmp = value[0]/value[1]
    ave_eui.append(tmp)
type_ave_eui = eui_count.copy()
type_ave_eui["ave_eui"] = ave_eui
type_ave_eui = type_ave_eui.loc[:,["ave_eui"]]
type_ave_eui = type_ave_eui.sort_values(by='ave_eui', ascending=False)
type_ave_eui
```

```python
type_ave_eui.plot.bar(figsize=(15,5),title="Average EUI for Main Types")
```
