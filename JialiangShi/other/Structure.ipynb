```python
import csv
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("together.csv",encoding='latin-1')
```

```python
df
```

```python
structure = df[['STRUCTURE_CLASS','EUI']]
structure
```

```python
structure = structure[structure['STRUCTURE_CLASS'].notnull()]
structure
```

```python
structure["count"] = np.ones(len(structure))
count = structure.groupby("STRUCTURE_CLASS").sum()
count
```

```python
count["average"] = count["EUI"] / count["count"]
count['average']
```

```python
count = count.loc[:,["average"]]
```

```python
count
```

```python
%matplotlib inline
count.plot.bar(figsize=(10,5),title="How structure affects EUI")
```

```python
df_HDD=pd.read_csv("KBOS_HDD_65F-3.csv",encoding='latin-1')
missing_HDD = [1203,1282,980]

hdd_2015 = sum(missing_HDD) + sum(df_HDD['HDD'][0:9])
hdd_2016 = sum(df_HDD['HDD'][9:21])
hdd_2017 = sum(df_HDD['HDD'][21:33])
```

```python
df_CDD=pd.read_csv("KBOS_CDD_65F-3.csv",encoding='latin-1')
df_CDD
```

```python
cdd_2015 = sum(df_CDD['CDD'][0:9]) + sum([0,0,0])
cdd_2016 = sum(df_CDD['CDD'][9:21])
cdd_2017 = sum(df_CDD['CDD'][21:33])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_axis = ['2015','2016','2017']
x = range(len(x_axis))
y_cdd = [cdd_2015,cdd_2016,cdd_2017]
y_hdd = [hdd_2015,hdd_2016,hdd_2017]
plt.plot(x, y_cdd, 'bo-')
plt.xticks(x, x_axis, rotation=45)
plt.margins(0.08)
plt.subplots_adjust(bottom=0.15)
plt.title('2015-2017 CDD')
plt.ylabel('CDD')
plt.show()
```

```python
plt.plot(x, y_hdd, 'ro-')
plt.xticks(x, x_axis, rotation=45)
plt.margins(0.08)
plt.subplots_adjust(bottom=0.15)
plt.title('2015-2017 HDD')
plt.ylabel('HDD')
plt.show()
```
