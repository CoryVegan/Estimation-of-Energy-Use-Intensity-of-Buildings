```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("together.csv",encoding='latin-1')
```

```python
structure = df[['STRUCTURE_CLASS','EUI']]
structure = structure[structure['STRUCTURE_CLASS'].notnull()]
structure
```

```python
structure.sort_values(by='EUI',ascending=False)
```

```python
to_drop = []
for key,value in structure.iterrows():
    if value[1]>800:
        to_drop.append(key)
structure = structure.drop(to_drop)
structure.sort_values(by='EUI', ascending=False)
```

```python
structure_copy = structure.copy()
structure_copy["count"] = np.ones(len(structure),dtype=int)
count = structure_copy.groupby("STRUCTURE_CLASS").sum()
count
```

```python
count["ave_eui"] = count["EUI"] / count["count"]
count = count.sort_values(by='ave_eui', ascending=False).iloc[:,[2]]
count
```

```python
x = ["A","B","C","D","E","R"]
y = ["Struct Steel","Reinforced Concrete","Brick/Concrete","Wood/Frame","Metal","Residential"]
dict1 = dict(zip(x, y))
```

```python
tmp = count.copy()
tmp.index = [dict1[i] for i in list(count.index)]
tmp
```

```python
tmp.plot(kind='bar',figsize=(8,5),title="How structure classes affect average EUI", fontsize=13)
```

```python
df_HDD=pd.read_csv("KBOS_HDD_65F-3.csv",encoding='latin-1')
missing_HDD = [1203,1282,980]

hdd_2015 = sum(missing_HDD) + sum(df_HDD['HDD'][0:9])
hdd_2016 = sum(df_HDD['HDD'][9:21])
hdd_2017 = sum(df_HDD['HDD'][21:33])
```

```python
df_CDD=pd.read_csv("KBOS_CDD_65F-3.csv",encoding='latin-1')
df_CDD
```

```python
cdd_2015 = sum(df_CDD['CDD'][0:9]) + sum([0,0,0])
cdd_2016 = sum(df_CDD['CDD'][9:21])
cdd_2017 = sum(df_CDD['CDD'][21:33])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_axis = ['2015','2016','2017']
x = range(len(x_axis))
y_cdd = [cdd_2015,cdd_2016,cdd_2017]
y_hdd = [hdd_2015,hdd_2016,hdd_2017]
plt.plot(x, y_cdd, 'bo-')
plt.xticks(x, x_axis, rotation=45)
plt.margins(0.08)
plt.subplots_adjust(bottom=0.15)
plt.title('2015-2017 CDD')
plt.ylabel('CDD')
plt.show()
```

```python
plt.plot(x, y_hdd, 'ro-')
plt.xticks(x, x_axis, rotation=45)
plt.margins(0.08)
plt.subplots_adjust(bottom=0.15)
plt.title('2015-2017 HDD')
plt.ylabel('HDD')
plt.show()
```
