```python
import csv
import pandas as pd
import numpy as np
```

```python
berdo=pd.read_csv("berdo2017.csv",encoding='utf-8')
berdo
```

```python
data = berdo[['Address','ZIP','Site EUI (kBTU/sf)','Property Type']]
```

```python
# clear out the "not available" data
data = data[data['Site EUI (kBTU/sf)']!='Not Available']
# clear out the NaN
data = data[data['Site EUI (kBTU/sf)'].notnull()]
data_house = data[data['Property Type']=='Multifamily Housing']
data_office = data[data['Property Type']=='Office']
```

```python
assess=pd.read_csv("property-assessment-fy2017.csv",encoding='latin-1')
```

```python
# find the important information for matching
assess[['ST_NUM','ST_NAME','ST_NAME_SUF','ZIPCODE']]
```

```python
# check how many type of ST_NAME_SUF it has
st_type = set(assess['ST_NAME_SUF'])
```

```python
result = []
single_result = []
for idx, row in data_office.iterrows():
    #num,name,st_type = row['Address'].split()
    info = row['Address'].split()
    # make sure name are all in lower case
    info = [item.lower() for item in info]  
    if 'ave' in info or 'avenue' in info:
        AV = assess[assess['ST_NAME_SUF']=='AV']
        for index, row_av in AV.iterrows():
            if row_av['ST_NAME'].lower() in info and row_av['ST_NUM'].lower() in info:
                for i in AV.loc[index]:
                    single_result.append(i)
                for i in data.loc[idx]:
                    single_result.append(i)
                result.append(single_result)
                single_result = []
                break
                
    elif 'st' in info or 'street' in info or 'st.' in info:
        ST = assess[assess['ST_NAME_SUF']=='ST']
        for index, row_av in ST.iterrows():
            if row_av['ST_NAME'].lower() in info and row_av['ST_NUM'].lower() in info:
                for i in ST.loc[index]:
                    single_result.append(i)
                for i in data.loc[idx]:
                    single_result.append(i)
                result.append(single_result)
                single_result = []
                break
    else:
        Other = assess[assess['ST_NAME_SUF']!='ST']
        Other = Other[Other['ST_NAME_SUF']!='AV']
        for index, row_av in Other.iterrows():
            if row_av['ST_NAME'].lower() in info and row_av['ST_NUM'].lower() in info:
                for i in Other.loc[index]:
                    single_result.append(i)
                for i in data.loc[idx]:
                    single_result.append(i)
                result.append(single_result)
                single_result = []
                break
```

```python
result = np.array(result)
```
