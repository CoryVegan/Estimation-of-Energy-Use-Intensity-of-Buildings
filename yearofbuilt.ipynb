```python
import csv
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("together.csv",encoding='latin-1')
df = df[np.isnan(df['YR_BUILT']) == False]
df
```

```python
LU=[]
for row in df.itertuples():
    if row.LU not in LU:
        LU.append(row.LU)
LU
```

```python
#C
df_C=df[df['LU']=='C']
df_C
```

```python

eui={}
#dic.setdefault(key，[]).append(value)
for row in df_C.itertuples():
        if row.YR_BUILT==0.0:
            continue
        if row.YR_BUILT in eui:
            eui[row.YR_BUILT].append(row.EUI)
        else:
            eui[row.YR_BUILT]=[]
            eui[row.YR_BUILT].append(row.EUI)
for key in eui:
    sum_eui=0
    for num in eui[key]:
        sum_eui+=num
        ave=sum_eui/len(eui[key])
    eui[key]=ave
```

```python

new={}
keys=sorted(eui.keys())
for key in keys:
    new[key]=eui[key]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,6))
YEAR_C=[]
EUI_C=[]
for key in new:
    YEAR_C.append(key)
    EUI_C.append(eui[key])

plt.plot(YEAR_C, EUI_C) 
plt.title('Type C:Commercial')
plt.xlabel('Year of Built')
plt.ylabel('EUI')
plt.show()
```
